```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import os
from sklearn.model_selection import train_test_split
import PIL
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```python
DATADIR='../input/ieee-indian/Indian Currencies'
CATEGORIES=['10','100','20','200','2000','50','500']

IMG_SIZE=250
```

```python
training_data=[]
def create_training_data():
    for category in CATEGORIES:
        path=os.path.join(DATADIR, category)
        class_num=CATEGORIES.index(category)
        for img in os.listdir(path):
            try:
                img_array=cv2.imread(os.path.join(path,img))   #,cv2.IMREAD_GRAYSCALE)
                new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
                training_data.append([new_array,class_num])
            except Exception as e:
                pass
create_training_data()   
```

```python
df=pd.DataFrame(training_data)
```

```python
import random
random.shuffle(training_data)
```

```python
X=[]
y=[]

for categories, label in training_data:
    X.append(categories) 
    y.append(label)
xx= np.array(X).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
```

```python
X=np.array(X)
```

```python
X.shape
```

```python
plt.imshow(X[0])
print(CATEGORIES[y[0]])
print(len(X))
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
#x_train=np.array(x_train)
y_train=np.array(y_train)
```

```python
model = Sequential([
  layers.Rescaling(1./255, input_shape=(IMG_SIZE,IMG_SIZE,3)),
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(7,activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=10)
```

```python
ypred=model.predict(x_test)
```

```python
y_pri=[]
for i in ypred:
    y_pri.append(np.argmax(i))
    
```

```python
y_pri=np.array(y_pri)
```

```python
def imsh_test(i):
    plt.imshow(x_test[i])
    print(CATEGORIES[y_pri[i]])
```
